```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
labels = test['Activity']
encoder.fit(labels)

Label_encode = encoder.transform(labels)

Label_encode
```

```python
list(encoder.classes_)
```

```python
Activity_Label = pd.DataFrame(Label_encode, columns=['Label'])
type(Activity_Label)
```

```python
n_test = pd.concat([test, Activity_Label], axis=1)
```
