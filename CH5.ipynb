##### 5-7-3

```python
# a : b ； i = 0, 2, 4, 6, 8...
sec1 = {chr(ord('a')+i):chr(ord('a')+i+1) for i in range(26) if i % 2 == 0}
print(sec1)

# b : a ； i = 1, 3, 5, 7, 9...
sec2 = {chr(ord('a')+i):chr(ord('a')+i-1) for i in range(26) if i % 2 }
print(sec2)
```

```python
# 字典合併
sec1.update(sec2)
sec1
```

```python
s = 'an apple a day keeps the doctor away'
```

```python
# 進行加密
ss = ''
for c in s:
    if c != ' ':
        ss += sec1[c]
    else :
        ss += ' '
ss
```

```python
# 進行解密
us = ''
for c in ss:
    if c != ' ':
        us += sec1[c]
    else :
        us += ' '
us
```

### 1 求平方和

```python
n = int(input('請輸入n值：'))
sqsum = 0
for i in range (1, n+1) :
    sqsum += i**2
print(sqsum)
```

### 2 列出1~1000不被2與3整除

```python
for i in range (1, 1000+1):
    if(i%2 and i%3):
        print(i)
```

### 3 求大於1000最小平方和

```python
n = 0
sqsum = 0
while (sqsum < 1000):
    n +=1
    sqsum += n**2
print("最小的 n 值為 ", n)
```

### 4 19\*19 乘法表

```python
for a in range (1,20):
    for b in range (1,20):
        print(a, '*', b, '=', a*b, '\t', sep='', end='')
    print()
    print()
```

### 5 完全數

```python
for i in range (2,1001):
    i_sum = 0
    for j in range (1, i):
        if( i % j == 0):
            i_sum += j
    if( i == i_sum ):
        print(i, ' 為完全數')
```
